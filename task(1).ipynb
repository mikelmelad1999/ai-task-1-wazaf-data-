```python
# import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (1) Read the dataset, convert it to DataFrame and display some from it. 
df = pd.read_csv("Wuzzuf_Jobs.csv")
df.head()
```

```python
# (2) Display structure and summary of the data. 
df.describe()
```

```python
# (3) Clean the data (null, duplications) 
df.isnull().sum()

#df.dropna(inplace =True) to remove null values if it exists
```

```python
sum(df.duplicated())
```

```python
df.drop_duplicates(inplace =True)
sum(df.duplicated())
```

```python
# (4) Count the jobs for each company and display that in order (What are the most demanding companies for jobs?) 
demanding_companies = df["Company"].value_counts()
demanding_companies
```

```python
c= demanding_companies[:10].index
values = demanding_companies[:10].values
plt.pie(values ,labels =c )
plt.figure(figsize=(50, 50))
plt.show();
```

```python
# (5) Show step 4 in a pie chart 
c= demanding_companies[:10].index
values = demanding_companies[:10].values
plt.title("number of jobs for each company" ,fontsize=25)
plt.xlabel("company" ,fontsize =20)
plt.ylabel ("Number OF Jobs" ,fontsize =20)
plt.xticks(rotation=90)
plt.bar(c,values )
plt.figure(figsize=(.5, .5))
plt.show();
```

```python
# (6) Find out what are the most popular job titles. 
jobs =df["Title"].value_counts()
jobs 
```

```python
jobss= jobs.index[:10]
values = jobs.values [:10]
plt.bar (jobss ,values )
plt.title("The Most Popular Job Titles" ,fontsize=25)
plt.xlabel("Jobs" ,fontsize =20)
plt.ylabel ("Number OF Jobs" ,fontsize =20)
plt.xticks(rotation=90)
plt.show()
```

```python
# (7) Show step 6 in pie chart 
jobss= jobs.index[:10]
values = jobs.values [:10]
plt.pie(values ,labels =jobss )
plt.show()
```

```python
# (8) Find out the most popular areas? 
areas =df["Location"].value_counts()
areas
```

```python
c = areas.index[:10]
values = areas.values [:10]
plt.bar (c ,values )
plt.title("The Most Popular Areas" ,fontsize=25)
plt.xlabel("Areas" ,fontsize =20)
plt.ylabel ("Number" ,fontsize =20)
plt.xticks(rotation=90)
plt.show()
```

```python
# (9) Show step 8 in pie chart 
c = areas.index[:10]
values = areas.values [:10]
plt.pie(values ,labels =c )
plt.show()
```

```python
# (10) Print skills one by one and how many each repeated and order the output to find out the most important skills required? 
sk =df["Skills"].str.cat(sep =",")
sk_count =pd.Series (sk.split(",")).value_counts()
sk_count
```

```python
c = sk_count.index[:10]
values = sk_count.values [:10]
plt.bar (c ,values )
plt.title("The Most Popular Skills" ,fontsize=25)
plt.xlabel("Skill" ,fontsize =20)
plt.ylabel ("Number" ,fontsize =20)
plt.xticks(rotation=90)
plt.show()
```
